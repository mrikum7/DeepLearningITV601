```python
import numpy as np
from tensorflow import keras
import tensorflow as tf
from PIL import Image
import matplotlib.pyplot as plt
import os
from skimage.color import rgb2gray
from keras.utils import to_categorical
```

```python
X_train = []
for j in range(9):
    for i, file in enumerate(os.listdir('../Datasets/xPLNET/DATASET/Training Samples/' + str(j))):
        if i < 100:
            data = Image.open(f'../Datasets/xPLNET/DATASET/Training Samples/{j}/' + file)
            data = np.asarray(data)
            X_train.append(rgb2gray(data))
```

```python
X_train = np.asarray(X_train)
```

```python
X_train = X_train.reshape((900, 64, 64, 1))
```

```python
y_train = []
for i in range(9):
    for j in range(100):
        y_train.append(i)
```

```python
y_train = to_categorical(y_train)
```

```python
y_train.shape
```

```python
X_train.shape
```

```python
model = keras.Sequential()
```

```python
model.add(keras.layers.Conv2D(64, (3, 3), activation = 'relu',
                              input_shape = (64, 64, 1)))
```

```python
#model.add(keras.layers.Conv2D(64, (3, 3), activation = 'relu'))
```

```python
model.add(keras.layers.MaxPooling2D((2, 2)))
```

```python
#model.add(keras.layers.Conv2D(32, (3, 3), activation = 'relu'))
```

```python
#model.add(keras.layers.MaxPooling2D((2, 2)))
```

```python
model.add(keras.layers.Conv2D(32, (3, 3), activation = 'relu'))
```

```python
model.add(keras.layers.MaxPooling2D((2, 2)))
```

```python
model.add(keras.layers.Flatten())
```

```python
model.add(keras.layers.Dense(100, activation = 'relu'))
```

```python
model.add(keras.layers.Dense(100, activation = 'relu'))
```

```python
model.add(keras.layers.Dense(9, activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs = 10, batch_size = 64)
```

## Testing Results

```python
X_test = []
for j in range(9):
    for i, file in enumerate(os.listdir('../Datasets/xPLNET/DATASET/Test Samples/' + str(j))):
        if i < 50:
            data = Image.open(f'../Datasets/xPLNET/DATASET/Test Samples/{j}/' + file)
            data = np.asarray(data)
            X_test.append(rgb2gray(data))
```

```python
X_test = np.asarray(X_test)
```

```python
X_test = X_test.reshape(450, 64, 64, 1)
```

```python
y_test = []
for i in range(9):
    for j in range(50):
        y_test.append(i)
```

```python
y_test = to_categorical(y_test)
```

```python
res = model.evaluate(X_test, y_test)
```
